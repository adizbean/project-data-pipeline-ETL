# A. Extract

```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import requests 
import time 
```

```python
# scraping data dari web Tinkerlust
driver = webdriver.Chrome()

image_products = []
product_brands = []
product_names = []
product_prices = []
for page in range (1,4): # mengambil data dari 3 page
    web_link = f'https://www.tinkerlust.com/wanita/tas?sort_by=live_date&sort_dir=desc&page=1&per_page=24{page}'
    driver.get(web_link)
    time.sleep(15)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    petak_list = soup.find_all('div', {'class':'pb-3 col'}) # merupakan code untuk satuan barang sehingga ketika pandas dijalankan dapat mengambil seluruh informasi barang
    for petak in petak_list:
        try:
            image_products.append(petak.find('img')['src']) # karena dalam code hmtl pada bagian image tidak memiliki 'attribute' serta 'attribute value', maka saya langsung meletakan 'src' pada syntax sehingga dapat langsung mengambil elemen gambar.
        except:
            image_products.append(None)
        try:
            product_brands.append(petak.find('div', {'class':'text-uppercase text-truncate font-weight-bold'}).get_text())
        except:
            product_brands.append(None)
        try:
            product_names.append(petak.find('div', {'class':'text-capitalize small pb-2'}).get_text())
        except:
            product_names.append(None)
        try:
           product_prices.append(petak.find_all('div', {'class': 'font-weight-bold'})[-1].get_text()) # karena code harga sama pada sebagian code brand, maka saya menambahkan '[-1]' supaya syntax dapat mengambil elemen terakhir yaitu harga barang.
        except: 
            product_prices.append(None)
            
driver.close()
```

```python
# memeriksa jumlah data pada tiap kolom jumlahnya sama
print(len(image_products))
print(len(product_brands))
print(len(product_names))
print(len(product_prices))
```

```python
df = pd.DataFrame()
```

```python
# menampilkan data dan menyesuaikan data dengan kolom
df['produk'] = product_names
df['brand'] = product_brands
df['harga'] = product_prices
df['gambar'] = image_products
```

```python
# menyimpan data kotor pada csv
df.to_csv('data_kotor.csv')
```

# B. Transform

```python
import pandas as pd
```

```python
# menambah "index_col=0" supaya pandas dapat membaca index yg sudah ada
df_bersih = pd.read_csv("data_kotor.csv", index_col=0)
```

```python
# meubah data pada kolom harga supaya tidak terdapat karakter lain selain angka
df_bersih["harga"] = df_bersih["harga"].str.replace("IDR", "")
df_bersih["harga"] = df_bersih["harga"].str.replace(",", "")
```

```python
# begitu pula pada data kolom produk, menghilangkan karakter "#" sehingga hanya menampilkan angka saja
df_bersih["produk"] = df_bersih["produk"].str.replace("#", "")
```
